```python
#open the prepared dataset (1981/12/15-2019/2/15)
data = xr.open_dataset('/glade/work/molina/DATA/jan2017_synoptic/case_study_ssts.nc', decode_cf=True)
```

```python
#########################################################################################
#########################################################################################
#########################################################################################


for counter, numerate in enumerate(np.repeat(np.arange(0,3,1),2)):
    
    iterate = "{:02d}".format(counter)

    fig = plt.figure(figsize=(6.,4.))
    ax = fig.add_axes([0., 0., 1., 1.])

    ax.set_title('SSTAs', fontsize=16)
    ax.text(0.9, 1.015,'Jan '+pd.to_datetime(data['T'].values[numerate]).strftime("%d"), 
            fontsize=16,
            transform=ax.transAxes)

    ax.pcolormesh(data.sst[counter,:,:], cmap=plt.cm.get_cmap('seismic'), vmin=-3, vmax=3, shading='gouraud')
    
    #plt.savefig(f'/glade/work/.../ams_sstas_{iterate}.png', bbox_inches='tight', dpi=200)
    plt.show()
    plt.close()


#########################################################################################
#########################################################################################
#########################################################################################
```

```python
import subprocess

def grid2gif(image_str, output_gif):
    str1 = 'convert -delay 8 -loop 0 ' + image_str  + ' ' + output_gif
    subprocess.call(str1, shell=True)

grid2gif("/glade/work/.../ams_sstas_{00..03}.png", "/glade/scratch/.../ams_sstas.gif")
```
